```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```



*   step 2:Create the data frame




```python
dataframe = pd.read_csv("/content/Zomato data .csv")
print(dataframe )
```

# convert the data type of column-rate

```python
def handleRate(value):
  valu = str(value).split('/')
  value = value[0];
  return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Type of resturant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturant")
```

# conclusion - majority of the resturant falls in dinning category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="blue", marker="o")
plt.xlabel("Type of resturant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
plt.xlabel("Type of resturant", c="red",size=20)
plt.ylabel("Votes", c="red", size=20)
```

# conclusion - dinning resturant has recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=3)
plt.title("rating distribution")
plt.show()
```

# conclusion - the majority resturant received ratings from 3.5 to 4

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# conclusion- the majority of couples preferr resturant with an approximate cost of 300 Rs

```python
dataframe.head()
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(x= 'online_order',y='rate',data=dataframe)
```

# conclusion - offline order received lower rating in comparison to online order

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel("online order")
plt.ylabel("Listed In (Type)")
plt.show()
```
